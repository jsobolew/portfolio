```python
import numpy as np
import pandas as pd
import yfinance as yf
import datetime as dt

import matplotlib.pyplot as plt
```

```python
inflation = pd.read_csv("EU_Inflation_HICP_data.csv")
inflation['date'] = inflation['Unnamed: 0']
inflation.drop(columns='Unnamed: 0', inplace=True)
inflation = inflation[['date','Poland',]]

WIG20 = pd.read_csv("wig20_m.csv")
month = dict({1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'Jun', 7:'Jul', 8:'Aug', 9:'Sep', 10:'Sep', 11:'Nov', 12:'Dec'})
for i in range(len(WIG20)):
    date = dt.datetime.strptime(WIG20['Data'][i], "%Y-%m-%d")
    newdate = str(date.year)+str(month[date.month])
    WIG20.loc[i,'date'] = newdate

WIG20 = WIG20[['date','Zamkniecie']]

data = pd.merge(WIG20, inflation[['date','Poland']], on='date')
data = data.iloc[140:]
data
```

```python
plt.figure(figsize=(20,10))
plt.plot(data['Poland'])
plt.plot(data['Zamkniecie']/500)
```
