```python
import myTA
import numpy as np
import pandas as pd
import yfinance as yf
import datetime as dt
import matplotlib.pyplot as plt
```

```python
tickers = pd.read_csv("nasdaqTickers.csv")
```

```python
results = pd.DataFrame(columns=['ticker', 'start', 'end', 'strategy', 'comission', 'result [$]','return [%]'])
start = dt.datetime(2022,12,1)
end = dt.datetime.now()
commission = 0

for i in range(10):
    ticker = tickers['Symbol'][i]
    try:
        data = myTA.downloadAndMakeAnalysis(ticker, start, end,'5m')
    except:
        pass

    simpleMACDNoLossOnTrade_result = myTA.calculateProfit(myTA.TradingStrategies.simpleMACDNoLossOnTrade(data), commission)
    results = pd.concat([results, pd.DataFrame({'ticker':[ticker], 'start':[start], 'end':[end], 'strategy':['simpleMACDNoLossOnTrade'], 'comission':[commission],
     'result [$]':[simpleMACDNoLossOnTrade_result], 'return [%]':[simpleMACDNoLossOnTrade_result/data['Open'][0]]})])

    bestPossible_result = myTA.calculateProfit(myTA.TradingStrategies.bestPossible(data), commission)
    results = pd.concat([results, pd.DataFrame({'ticker':[ticker], 'start':[start], 'end':[end], 'strategy':['bestPossible'], 'comission':[commission],
     'result [$]':[bestPossible_result], 'return [%]':[bestPossible_result/data['Open'][0]]})])

    simpleMACD_result = myTA.calculateProfit(myTA.TradingStrategies.simpleMACD(data), commission)
    results = pd.concat([results, pd.DataFrame({'ticker':[ticker], 'start':[start], 'end':[end], 'strategy':['simpleMACD'], 'comission':[commission],
     'result [$]':[simpleMACD_result], 'return [%]':[simpleMACD_result/data['Open'][0]]})])

    simpleMACDZeroHistPrediction_result = myTA.calculateProfit(myTA.TradingStrategies.simpleMACDZeroHistPrediction(data), commission)
    results = pd.concat([results, pd.DataFrame({'ticker':[ticker], 'start':[start], 'end':[end], 'strategy':['simpleMACDZeroHistPrediction'], 'comission':[commission],
     'result [$]':[simpleMACDZeroHistPrediction_result], 'return [%]':[simpleMACDZeroHistPrediction_result/data['Open'][0]]})])

    buyAndHold_result = myTA.calculateProfit(myTA.TradingStrategies.buyAndHold(data), commission)
    results = pd.concat([results, pd.DataFrame({'ticker':[ticker], 'start':[start], 'end':[end], 'strategy':['buyAndHold'], 'comission':[commission],
     'result [$]':[buyAndHold_result], 'return [%]':[buyAndHold_result/data['Open'][0]]})])
```
