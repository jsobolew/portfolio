```python
import matplotlib.pyplot as plt
from autograd import grad
import numpy as np
from cec2017.functions import f1, f2, f3
```

Funkcja Booth

```python
def booth(x):
    return (x[0] + 2*x[1] - 7)**3 + (2*x[0]+x[1]-5)**2
```

```python
UPPER_BOUND = 100
DIMENSIONALITY = 2
x = np.random.uniform(-UPPER_BOUND, UPPER_BOUND, size=DIMENSIONALITY)
beta = 0.000001
f = booth
loss = []
xhistory = []
d = grad(f)

for i in range(10000):
    x = x-beta*d(x)
    loss.append(f(x))
    xhistory.append(x)

plt.plot(loss)
```

```python
UPPER_BOUND = 100
DIMENSIONALITY = 2
x = np.random.uniform(-UPPER_BOUND, UPPER_BOUND, size=DIMENSIONALITY)
beta = 0.00000001
f = booth
d = grad(f)

MAX_X = 100
PLOT_STEP = 0.1

x_arr = np.arange(-MAX_X, MAX_X, PLOT_STEP)
y_arr = np.arange(-MAX_X, MAX_X, PLOT_STEP)
X, Y = np.meshgrid(x_arr, y_arr)
Z = np.empty(X.shape)

for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        Z[i, j] = f(np.array([X[i, j], Y[i, j]]))
        
plt.contour(X, Y, Z, 500)
for i in range(len(xhistory)-1):
    plt.arrow(xhistory[i][0], xhistory[i][1], xhistory[i+1][0] - xhistory[i][0], xhistory[i+1][1] - xhistory[i][1], head_width=1, head_length=1, fc='k', ec='k')
plt.show()
```
