```python
from ID3 import ID3
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
def test(max_depth, iters, X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)
    accuracy_train = np.zeros((max_depth-1,))
    accuracy_val = np.zeros((max_depth-1,))

    for i in range(iters):
        for depth in range(1, max_depth):
            model = ID3(max_depth = depth)
            model.fit(X_train, y_train)

            preds = model.predict(X_test)
            accuracy_val[depth-1] += accuracy_score(preds, y_test)
            preds_train = model.predict(X_train)
            accuracy_train[depth-1] += accuracy_score(preds_train, y_train)
    print(confusion_matrix(y_test, preds))
    accuracy_val /= iters
    accuracy_train /= iters

    plt.plot(accuracy_val)
    plt.plot(accuracy_train)
    plt.xlabel('depth')
    plt.ylabel('accuracy')
    plt.legend(['val','train'])
```

```python
breast = pd.read_csv("breast-cancer.csv")
X = breast.drop(columns='Class')
y = breast['Class'].values

test(30, 10, X, y)
```

```python
schrooms = pd.read_csv("agaricus-lepiota.csv")
X = schrooms.drop(columns='Class')
y = schrooms['Class'].values

test(30, 10, X, y)
```
