```python
from pyspark.ml.feature import StringIndexer, OneHotEncoder, VectorAssembler, StandardScaler
from pyspark.ml import Pipeline
from pyspark.sql.functions import col

spark_df= spark.sql(f'SELECT *, CAST((convertedComp > 60000) AS STRING) AS compAboveAvg \
                    FROM {table_name} where convertedComp IS NOT NULL ')

numerical_features = ["YearsCode"]
# numericalFeature_stages = [spark_df.withColumn(c+"Num", col(c).cast("int")) for c in numerical_features]
spark_df = spark_df.withColumn("YearsCode",col("YearsCode").cast("double"))
# numericalFeature_stages = [StandardScaler(inputCol=c, outputCol='feature_' + c) for c in numerical_features]
# chcemy przewidziec compAboveAvg
y = 'compAboveAvg'
# na podstawie:
feature_columns = ['OpSys', 'EdLevel', 'MainBranch' , 'Country', 'JobSeek', 'YearsCode', 'OrgSize', 'WorkWeekHrs', 'YearsCodePro', 'NEWLearn', 'JobSat', 'Gender', 'CurrencySymbol', 'Age1stCode']

# Refactoring do list comprehension
stringindexer_stages = [StringIndexer(inputCol=c, outputCol='stringindexed_' + c).setHandleInvalid("keep") for c in feature_columns]

# i dla zmiennej objaśnianej
stringindexer_stages += [StringIndexer(inputCol=y, outputCol='label')]

onehotencoder_stages = [OneHotEncoder(inputCol='stringindexed_' + c, outputCol='feature_' + c) for c in feature_columns]

pa = Pipeline(stages=stringindexer_stages + onehotencoder_stages).fit(spark_df).transform(spark_df).toPandas()
```

```python
extracted_columns = ['feature_' + c for c in feature_columns] #+ ['YearsCode']
vectorassembler_stage = VectorAssembler(inputCols=extracted_columns, outputCol='features') 

final_columns = [y] + feature_columns + extracted_columns + ['features', 'label']

transformed_df = Pipeline(stages=stringindexer_stages + \
                          onehotencoder_stages + \
                          # numericalFeature_stages + \
                          [vectorassembler_stage]).fit(spark_df).transform(spark_df).select(final_columns)

transformed_df.limit(5).toPandas()
```

# Podział na testyowy i treningowy

```python
training, test = transformed_df.randomSplit([0.8, 0.2], seed=1234)

print(f"train: {training.count()} test: {test.count()}")
```

```python
from pyspark.ml.classification import GBTClassifier
gbt = GBTClassifier(labelCol="label", featuresCol="features", maxIter=10)
model = gbt.fit(training)
```

```python
# Ewaluator 
from pyspark.ml.evaluation import BinaryClassificationEvaluator
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
AUROC = BinaryClassificationEvaluator(rawPredictionCol="rawPrediction", metricName="areaUnderROC")
accuracy = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="accuracy")

print(f"AUROC: {AUROC.evaluate(model.transform(test))} accuracy: {accuracy.evaluate(model.transform(test))}")
```

```python
# Confusion matrix
pred = model.transform(test)
label_and_pred = pred.select('label', 'prediction')
label_and_pred.groupBy('label', 'prediction').count().toPandas()
```
