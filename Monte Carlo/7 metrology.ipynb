```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def runMetrology(X1, X2):
    M = int(1e6)
    p = 0.9545

    x1, x2 =  X1(M), X2(M)
    yr = x1 + x2

    y_hat = np.mean(yr)

    sigma_y_hat = np.sqrt(1/(M-1)*np.sum(np.power(yr-y_hat, 2)))

    sigma_c = np.sqrt(np.var(x1)+np.var(x2))

    q = p*M
    yr = np.sort(yr)
    r_ = (M - q)/2

    print("MC -----------------")
    print(f"sigma {sigma_y_hat}")
    print(f"przedział [{yr[int(r_)]}, {yr[int(r_+q)]}]")
    print("\n")
    print("Teoria -----------------")
    print(f"sigma {sigma_c}")
    print(f"przedział [{-2*sigma_c}, {2*sigma_c}]")

    plt.figure(figsize=(20,5))
    plt.subplot(1,3,1)
    plt.hist(x1)
    plt.title("x1")

    plt.subplot(1,3,2)
    plt.hist(x2)
    plt.title("x2")

    plt.subplot(1,3,3)
    plt.hist(yr)
    plt.title("yr")
```

# zad 1

```python
def X1(n):
    return np.random.normal(0, 1, n)

def X2(n):
    return np.random.normal(0, 2, n)

runMetrology(X1, X2)
```

# zad 2

```python
def X1(n):
    return np.random.uniform(0, 4, n)

def X2(n):
    return np.random.uniform(5, 6, n)

runMetrology(X1, X2)
```
