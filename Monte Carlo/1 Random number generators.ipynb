```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
```

```python
def generator(Xn, a, c, m):
    return (a*Xn + c) % m
```

```python
number = 1
a, c, m = 17, 0, 2e13-1

genNumbers = []

for _ in range(2500):
    number = generator(number, a, c, m)
    genNumbers.append(number)


plt.plot(genNumbers[:-1], genNumbers[1:], '.')
plt.title(f"Xn z Xn+1 corr: {np.correlate(genNumbers[:-1], genNumbers[1:])[0]}")
plt.show()

plt.plot(genNumbers[:-2], genNumbers[2:], '.')
plt.title(f"Xn z Xn+2 corr: {np.correlate(genNumbers[:-2], genNumbers[2:])[0]}")
plt.show()
plot_acf(np.array(genNumbers))
plt.show()
```

Korelacja nastpnej generowanej liczby z poprzedni jest do spora (ukone pasy), natomiast z drug, nie jest widoczna. Podobn rzecz mona zauway na wykresie autokorelacji, pierwsza warto oczywiscie jest rwna 1, a nastpna jest dosy spora, a potem mniej lub bardziej oscyluje w ok zera. Liczba startowa nie ma znacznego wpywu na jakociowe wyniki generatora.

```python
offset = 26
plt.plot(genNumbers[:-offset], genNumbers[offset:], '.')
plt.show()
```

Warto te zwrci uwag, e wykres autokorelacji nie koniecznie jest dobrym wyznacznikiem jakoi generatora. Z wykresy wynika, e 26 liczba porwnywanie dobrze koreluje si z pierwsz, jednake to o niczym nie znaczy o niczym nie wiadczy. Wizualnie wygenerowane liczby wygldaj dobrze. Nie wida adnego wzoru.

```python
randomnumers = np.random.rand(2500)
plt.plot(randomnumers[:-1], randomnumers[1:], '.')
plt.show()

plot_acf(np.array(randomnumers))
plt.show()
```
