```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
```

```python
def generator(Xn, a, c, m):
    return (a*Xn + c) % m
```

```python
number = 1
a, c, m = 17, 0, 2e13-1

genNumbers = []

for _ in range(2500):
    number = generator(number, a, c, m)
    genNumbers.append(number)


plt.plot(genNumbers[:-1], genNumbers[1:], '.')
plt.title(f"Xn z Xn+1 corr: {np.correlate(genNumbers[:-1], genNumbers[1:])[0]}")
plt.show()

plt.plot(genNumbers[:-2], genNumbers[2:], '.')
plt.title(f"Xn z Xn+2 corr: {np.correlate(genNumbers[:-2], genNumbers[2:])[0]}")
plt.show()
plot_acf(np.array(genNumbers))
plt.show()
```

Korelacja następnej generowanej liczby z poprzednią jest dość spora (ukośne pasy), natomiast z drugą, nie jest widoczna. Podobną rzecz można zauważyć na wykresie autokorelacji, pierwsza wartość oczywiscie jest równa 1, a następna jest dosyć spora, a potem mniej lub bardziej oscyluje w okół zera. Liczba startowa nie ma znacznego wpływu na jakościowe wyniki generatora.

```python
offset = 26
plt.plot(genNumbers[:-offset], genNumbers[offset:], '.')
plt.show()
```

Warto też zwrócić uwagę, że wykres autokorelacji nie koniecznie jest dobrym wyznacznikiem jakośći generatora. Z wykresy wynika, że 26 liczba porównywanie dobrze koreluje się z pierwszą, jednakże to o niczym nie znaczy o niczym nie świadczy. Wizualnie wygenerowane liczby wyglądają dobrze. Nie widać żadnego wzoru.

```python
randomnumers = np.random.rand(2500)
plt.plot(randomnumers[:-1], randomnumers[1:], '.')
plt.show()

plot_acf(np.array(randomnumers))
plt.show()
```

Porównując do liczb losowych generowanych w Numpy. Nasz generator posiada większą korelację między liczbami.

```python
number = 10
a, c, m = 85, 0, 2e13-1

genNumbers = []

for _ in range(2500):
    number = generator(number, a, c, m)
    genNumbers.append(number)


plt.plot(genNumbers[:-1], genNumbers[1:], '.')
plt.title(f"Xn z Xn+1 corr: {np.correlate(genNumbers[:-1], genNumbers[1:])[0]}")
plt.show()

plt.plot(genNumbers[:-2], genNumbers[2:], '.')
plt.title(f"Xn z Xn+2 corr: {np.correlate(genNumbers[:-2], genNumbers[2:])[0]}")
plt.show()
plot_acf(np.array(genNumbers))
plt.show()
```

Dla wartośći a = 85 korelacja następnej generowanej liczny z poprzednią jest mniejsza niż dla a = 17, jendkże nadal jest ona widoczna gołym okiem. Dla 2 następnej liczy korelacja jest ponownie dosyć mała. Podobne wnioski można wyciągnąć na podstawie wykresu autokorelacji. Druga i trzecia wartość mają małą wartość, co świadczy o ich małej korelacji.
